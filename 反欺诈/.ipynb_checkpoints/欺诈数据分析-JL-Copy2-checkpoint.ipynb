# 寻找最优参数

```python

#li=[]
def run_model(X_train,Y_train,X_test, Y_test,layer2_kernel_l1,batch_size):
    #global li
    class RocAucEvaluation(Callback):
        def __init__(self, validation_data=(), interval=1):
            super(Callback, self).__init__()
            self.interval = interval
            self.x_val,self.y_val,self.layer2_kernel_l1,self.batch_size = validation_data
        def on_epoch_end(self, epoch, log={}):
            if epoch % self.interval == 0:
                #y_pred = self.model.predict(self.x_val, verbose=0)
                y_pred = self.model.predict(self.x_val, verbose=0)
                score = roc_auc_score(self.y_val, y_pred)
                s=str(score)+','+str(self.layer2_kernel_l1)+','+str(self.batch_size)
                #li.append([self.layer1_kernel_l1,self.layer1_kernel_l2,self.layer2_kernel_l1,self.layer2_kernel_l2,self.batch_size,score])
                with open('模型/dnn/一晚上.txt', "a") as f:
                    f.write(s+','+str(epoch)+'\n')
                if(score>0.8476):
                    model.save('模型/dnn/auc'+s+'.h5')
                    print('模型已保存:'+'模型/dnn/auc'+str(score)+'.h5')
                if(score<0.55):
                    raise 'AUC太低，别跑了'
                print('\n ROC_AUC - epoch:%d - score:%.6f \n' % (epoch+1, score))

    RocAuc = RocAucEvaluation(validation_data=(X_test, Y_test,
                                               layer2_kernel_l1,batch_size),
                              interval=1)
    X_train=X_train.values
    Y_train=Y_train.values
    X_test=X_test.values
    Y_test=Y_test.values
    model = Sequential()
    model.add(Dense(units=300, activation='relu', input_dim=200#,
                    #kernel_regularizer=regularizers.l2(0.01),
                    #activity_regularizer=regularizers.l1(0.01) #效果挺好
                    #kernel_regularizer=regularizers.l1(layer1_kernel_l1)#,
                   # activity_regularizer=regularizers.l1(layer1_kernel_l2)

                   ))
    model.add(Dense(units=200,activation='relu',
                    kernel_regularizer=regularizers.l1(layer2_kernel_l1)#,
                    #activity_regularizer=regularizers.l1(layer2_kernel_l2)

                   ))
    #model.add(Dense(units=64,activation='relu'))
    model.add(Dense(units=1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
    #model.compile(loss='binary_crossentropy',optimizer='adam',metrics=[roc_auc_score])
    #model.fit(X_train,Y_train,epochs=500,batch_size=128)
    try:
        model.fit(X_train, Y_train, batch_size=batch_size, epochs=100, validation_data=(X_test, Y_test), callbacks=[RocAuc], verbose=2)
    except:
        pass
    return model
    #evaluate the model
    '''scores = model.evaluate(X_test,Y_test)
    print("\n%s:%.2f%%" % (model.metrics_names[1],scores[1]*100))
    predictions=model.predict(X_test, batch_size=128)
    predictions=np.around(predictions,0)
    accruracy=accuracy_score(Y_test,predictions)
    cnf_matrix=confusion_matrix(Y_test,predictions)

    print(accruracy,cnf_matrix)
    roc_auc_score(Y_test,predictions)'''

    
#run_model(X_train,Y_train,X_test, Y_test,0.2034,64)
'''
layer1_kernel_l1_list=np.linspace(0.0,0,1)
layer2_kernel_l1_list=np.linspace(0.10,0.3,30)
batch_size_list=[32,64]

for layer1_kernel_l1 in layer1_kernel_l1_list:
    for layer2_kernel_l1 in layer2_kernel_l1_list:
        for batch_size in batch_size_list:
            print(layer1_kernel_l1,layer2_kernel_l1,batch_size)
            try:
                run_model(layer1_kernel_l1,layer2_kernel_l1,batch_size)
            except:
                pass
'''
```

# 对偏差数据再次训练

```python
model = load_model('模型/dnn/auc0.8511444998782415,0.0,0.20344827586206896,64.h5')
models=[]
models.append(model)
for i in range(10):
    x_train=X_train.copy()
    y_train=Y_train.copy()
    
    y=np.around(model.predict(X_test),0)
    last_pre=pd.DataFrame(x_train)
    x_train=x_train[y_train.iloc[:,0]!=last_pre.iloc[:,0]]
    y_train=y_train[y_train.iloc[:,0]!=last_pre.iloc[:,0]]
    model=run_model(x_train,y_train,X_test, Y_test,0.2034,64)
    models.append(model)
```
