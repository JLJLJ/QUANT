```python
from sklearn.metrics import roc_auc_score
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score  
from xgboost import XGBClassifier
import xgboost as xgb
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import Callback
from tensorflow.keras import regularizers
from sklearn.metrics import confusion_matrix
from tensorflow.keras.models import load_model
import numpy as np
from sklearn.model_selection import GridSearchCV
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras import optimizers
from tensorflow.keras.wrappers import scikit_learn
import lightgbm as lgb
from lightgbm import LGBMClassifier
from sklearn import model_selection
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB 
from sklearn.ensemble import RandomForestClassifier
from mlxtend.classifier import StackingClassifier
import mlxtend
import os
from tensorflow.keras import backend as K

# config = tf.ConfigProto()
# config.gpu_options.per_process_gpu_memory_fraction = 0.3
# K.set_session(tf.Session(config=config))
# os.environ['CUDA_VISIBLE_DEVICES']='0'

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

from sklearn.metrics import roc_auc_score

import matplotlib as mpl
import seaborn as sns

import time
from functools import wraps
#import pysnooper


class params_iter:
    def getPlans(self,lis,jude=True):
        if jude: 
            lis = [[[i] for i in lis[0]]] + lis[1:]
        if len(lis) > 2:
            for i in lis[0]:
                for j in lis[1]:
                    self.getPlans([[i + [j]]] + lis[2:], False)
        elif len(lis) == 2:
            for i in lis[0]:
                for j in lis[1]:
                    self.param_list.append(i + [j])
                
    def __init__(self,params):
        self.params=params
        self.cur_index=0
        self.param_list=[]
        val=list(params.values())
        keys=list(params.keys())
        self.getPlans(val)
        self.df=pd.DataFrame(self.param_list,columns=keys)
#         print(self.df)
        
    def __iter__(self):
        return self
 
    def __next__(self):
        if(self.cur_index>=len(self.df)):
            raise StopIteration
        x = self.df.iloc[self.cur_index,:]
        self.cur_index += 1
        return x.to_dict()
    
def fn_timer(function):
  @wraps(function)
  def function_timer(*args, **kwargs):
    t0 = time.time()
    result = function(*args, **kwargs)
    t1 = time.time()
    print ("Total time running %s: %s seconds" %
        (function.__name__, str(t1-t0)))
    return result
  return function_timer

def show_auc(model,X_test,Y_test):
    
    predictions=model.predict(X_test)
    print('auc:',roc_auc_score(Y_test,predictions))
    
    accruracy=accuracy_score(Y_test,np.around(predictions,0))
    cnf_matrix=confusion_matrix(Y_test,np.around(predictions,0))

    print('accruracy:',accruracy)
    print(cnf_matrix)

def add_models(model,name):
    models.append([model,name])

%matplotlib inline

dataset_train=pd.read_hdf('Dataset/1_StandardScaler.h5', key='TRAIN')
dataset_test=pd.read_hdf('Dataset/1_StandardScaler.h5', key='PREDICT')
# dataset_train = pd.read_hdf('Dataset/2_categorical_to_int.h5', key='TRAIN')
# dataset_test = pd.read_hdf('Dataset/2_categorical_to_int.h5', key='PREDICT')

dataset_x=dataset_train.drop(columns =['target','tran_id'])
dataset_y=pd.DataFrame(dataset_train['target'])

X_train,X_test,Y_train,Y_test=train_test_split(dataset_x,dataset_y,test_size=0.01,random_state=0)
# X_train=dataset_x
# Y_train=dataset_y

models=[]
scores=[]
```

```python

#li=[]
def run_model(X_train,Y_train,X_test, Y_test,layer2_kernel_l1,batch_size):
    #global li
    class RocAucEvaluation(Callback):
        def __init__(self, validation_data=(), interval=1):
            super(Callback, self).__init__()
            self.interval = interval
            self.x_val,self.y_val,self.layer2_kernel_l1,self.batch_size = validation_data
        def on_epoch_end(self, epoch, log={}):
            if epoch % self.interval == 0:
                #y_pred = self.model.predict(self.x_val, verbose=0)
                y_pred = self.model.predict(self.x_val, verbose=0)
                score = roc_auc_score(self.y_val, y_pred)
                s=str(score)+','+str(self.layer2_kernel_l1)+','+str(self.batch_size)
                #li.append([self.layer1_kernel_l1,self.layer1_kernel_l2,self.layer2_kernel_l1,self.layer2_kernel_l2,self.batch_size,score])
                with open('模型/dnn/一晚上.txt', "a") as f:
                    f.write(s+','+str(epoch)+'\n')
                print('\n ROC_AUC - epoch:%d - score:%.6f \n' % (epoch+1, score))

    RocAuc = RocAucEvaluation(validation_data=(X_test, Y_test,
                                               layer2_kernel_l1,batch_size),
                              interval=1)
    X_train=X_train.values
    Y_train=Y_train.values
    X_test=X_test.values
    Y_test=Y_test.values
    model = Sequential()
    model.add(Dense(units=300, activation='relu'
                    #kernel_regularizer=regularizers.l2(0.01),
                    #activity_regularizer=regularizers.l1(0.01) #效果挺好
                    #kernel_regularizer=regularizers.l1(layer1_kernel_l1)#,
                   # activity_regularizer=regularizers.l1(layer1_kernel_l2)

                   ))
    model.add(Dense(units=200,activation='relu',
                    kernel_regularizer=regularizers.l1(layer2_kernel_l1)

                   ))
    model.add(Dense(units=1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
    try:
        model.fit(X_train, Y_train, batch_size=batch_size, epochs=50, validation_data=(X_test, Y_test), callbacks=[RocAuc], verbose=2)
    except:
        pass
    return model
    #evaluate the model

dnn=run_model(X_train,Y_train,X_test, Y_test,0.2034,64)
```
