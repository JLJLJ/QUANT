```python
#不好使
import os
from sklearn.preprocessing import LabelEncoder
from sklearn.externals import joblib

def categorical_to_int_and_savefile(dataset,categorical_columns):
    dataset=dataset.copy()
    FILE_PATH='特征编码模型'
    if not os.path.exists(FILE_PATH):        
        os.makedirs(FILE_PATH)
    for column in categorical_columns:
        le=LabelEncoder().fit(dataset[column].astype('str'))
        joblib.dump(le, FILE_PATH+'/'+column)
        dataset[column]=le.transform(dataset[column].astype('str'))
    return dataset

def categorical_to_int(dataset,categorical_columns):
    dataset=dataset.copy()
    FILE_PATH='特征编码模型'
    for column in categorical_columns:
        le=joblib.load(FILE_PATH+'/'+column)
        dataset[column]=le.transform(dataset[column].astype('str'))
    return dataset
categorical_columns=['attr_68']
dataset_train=categorical_to_int_and_savefile(dataset_train,categorical_columns)
dataset_test=categorical_to_int_and_savefile(dataset_test,categorical_columns)
dataset_train[categorical_columns]=dataset_train[categorical_columns].astype('category')
dataset_train.to_hdf('Dataset/X2_categorical_to_int.h5', key='TRAIN',format="table")
dataset_test.to_hdf('Dataset/X2_categorical_to_int.h5', key='PREDICT',format="table")
```
