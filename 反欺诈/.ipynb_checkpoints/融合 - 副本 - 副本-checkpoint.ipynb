```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
from imblearn.over_sampling import ADASYN
from imblearn.over_sampling import SVMSMOTE
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score  
from xgboost import XGBClassifier
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import Callback
from tensorflow.keras import regularizers
from sklearn.metrics import confusion_matrix
from tensorflow.keras.models import load_model
import numpy as np
from sklearn.model_selection import GridSearchCV
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras import optimizers
from tensorflow.keras.wrappers import scikit_learn

import os
from tensorflow.keras import backend as K
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

from sklearn.metrics import roc_auc_score

import matplotlib as mpl
import seaborn as sns

import time
from functools import wraps
import matplotlib as mpl
import seaborn as sns

import time
from functools import wraps

import pysnooper
config = tf.ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction = 0.3
K.set_session(tf.Session(config=config))
os.environ['CUDA_VISIBLE_DEVICES']='0'

def fn_timer(function):
  @wraps(function)
  def function_timer(*args, **kwargs):
    t0 = time.time()
    result = function(*args, **kwargs)
    t1 = time.time()
    print ("Total time running %s: %s seconds" %
        (function.__name__, str(t1-t0)))
    return result
  return function_timer
def get_sample(dataset,target_0=0.33,target_1=0.33,random_state=None):
    dataset=dataset.copy()
    x0=dataset[dataset['target']==0]
    x1=dataset[dataset['target']==1]
    x0_selected=x0.sample(frac=target_0,random_state=random_state)
    x1_selected=x1.sample(frac=target_1,random_state=random_state)
    dataset=pd.concat([x0_selected,x1_selected]).sample(frac=1,random_state=random_state)
    y_selected=dataset['target']
    x_selected=dataset.drop(columns =['target'])
    return x_selected,y_selected


%matplotlib inline
dataset = pd.read_csv("Dataset/trainData/ANTIF_TRAIN_SET.csv",header=0,index_col=0)
dataset_bk=dataset.copy()
```

```python
params={
    'num_leaves':[12,14,16],
    'min_data_in_leaf':[40],
    'objective':['binary'],
    'max_depth':[13,15,17],
    'learning_rate':[0.02],
    'boosting':['gbdt'],
    'bagging_freq':[8],
    'bagging_fraction':[0.7,0.8,0.9],
    'feature_fraction':[0.7],
    'bagging_seed':[6,8,10],
    'reg_alpha':[3,5,7],
    'reg_lambda':[6,8,10,12,14],
    'random_state':[42],
    'metric':['auc'],
    'verbosity':[-1],
    'subsample':[1],
    'min_gain_to_split':[0.02],
    'min_child_weight':[20,25],
    'num_threads':[12]
}

class params_iter:
    def getPlans(self,lis,jude=True):
        if jude: 
            lis = [[[i] for i in lis[0]]] + lis[1:]
        if len(lis) > 2:
            for i in lis[0]:
                for j in lis[1]:
                    self.getPlans([[i + [j]]] + lis[2:], False)
        elif len(lis) == 2:
            for i in lis[0]:
                for j in lis[1]:
                    self.param_list.append(i + [j])
                
    def __init__(self,params):
        self.params=params
        self.cur_index=0
        self.param_list=[]
        val=list(params.values())
        keys=list(params.keys())
        self.getPlans(val)
        self.df=pd.DataFrame(self.param_list,columns=keys)
#         print(self.df)
        
    def __iter__(self):
        return self
 
    def __next__(self):
        if(self.cur_index>=len(self.df)):
            raise StopIteration
        x = self.df.iloc[self.cur_index,:]
        self.cur_index += 1
        return x.to_dict()
models=[]
scores=[]
lgb_train = lgb.Dataset(X_train,Y_train)
lgb_eval = lgb.Dataset(X_test,Y_test,reference=lgb_train)
for i,param in enumerate(params_iter(params)):
    print(i,param)
    model=lgb.train(param,lgb_train,num_boost_round=30000,valid_sets=lgb_eval,early_stopping_rounds=1000)
    models.append(model)
    scores.append(model.best_score['valid_0']['auc'])

    
#输出得分最高的前五个的得分、迭代次数、参数
sc=pd.Series(scores).sort_values(ascending =False)
for i in list(sc[0:5].index):
    print(sc[i],models[i].best_iteration,models[i].params)
    print('')
```
