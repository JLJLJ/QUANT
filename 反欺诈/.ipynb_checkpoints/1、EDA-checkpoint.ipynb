```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

import matplotlib as mpl
import seaborn as sns
%matplotlib inline
pd.set_option('display.max_columns',None) 
pd.set_option('expand_frame_repr', False)
```

```python
dataset = pd.read_csv("Dataset/trainData/ANTIF_TRAIN_SET.csv",header=0,index_col='tran_id')
dataset.describe()
```

```python
plt.rcParams['figure.figsize'] = (8.0, 4.0) # 设置figure_size尺寸
fig,axes = plt.subplots()
dataset.iloc[:,1:6].plot(kind='box',ax=axes)
axes.set_ylabel('values of tip_pct')
```

```python
plt.rcParams['figure.figsize'] = (160.0, 10.0) # 设置figure_size尺寸
fig,axes = plt.subplots()
dataset.iloc[:,1:].plot(kind='box',ax=axes)
axes.set_ylabel('values of tip_pct')
fig.savefig('箱线图.png') 
```

```python
print('target类别：%s'%set(dataset['target']))
print('正类数据占比:%f%%'%(len(dataset[dataset['target']==1])/len(dataset)*100))
```
